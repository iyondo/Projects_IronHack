```python
data = []
for ind in range(1,325):
	url=f'https://eu-citizen.science/project/{ind}'
	htlm = requests.get(url).content
	soup = BeautifulSoup(htlm,"lxml")
	try:
		name = soup.find_all('h1',class_='wood-color')
		nqme =name[0].text
	except:
		continue

	keywords = soup.find_all('a',class_='badge bg-my-darkGreen')
	key= [key.text for key in keywords]

	try:
		cooridnator = soup.find_all('h6')
		cor = cooridnator[0].find('a').text
	except:
		cor =' '
	row = [nqme, key, cor]
	data.append(row)
data

```

```python
colums = ['Project','Keywords','Coordinator']
table = pd.DataFrame(data, columns=colums)
table
```

```python
table.to_csv('projects_data.csv')
```
